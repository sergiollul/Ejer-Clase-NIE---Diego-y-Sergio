```r
df = read.csv("dfIne.csv", sep = ";", header = TRUE, fileEncoding = "latin1")

# tamaño df
dim(df)
```

```r
head(df)
```

```r
# Assuming your data frame is named 'df' and columns are separated
total_rows <- df %>%
  filter(Nacionalidad == "Total" & Pais_de_nacimiento == "Total") %>%
  nrow()

total_rows
```

```r
# Count rows where Pais_de_nacimiento contains "Extranjero"
extranjero_count <- df %>%
  filter(grepl("Extranjero", Pais_de_nacimiento)) %>%
  nrow()

extranjero_count
```

```r
unique(df$Nacionalidad)
```

```r
df <- df %>%
  filter(
    Pais_de_nacimiento == "Extranjero" &
      !Nacionalidad %in% c(
        'Total', 
        'País de la UE27_2020 sin España', 
        'País de la UE28 sin España', 
        'Otro país de la Unión Europea sin España', 
        'País de Europa menos UE27_2020', 
        'País de Europa menos UE28', 
        'Otro país del resto de Europa', 
        'Otro país de África', 
        'De América del Norte', 
        'De Centro América y Caribe', 
        'Otro país de Centro América y Caribe', 
        'Otro país de Sudamérica', 
        'Otro país de Asia', 
        'De Oceanía', 
        'Apátridas', 
        "De Africa", 
        "De Asia", 
        "De Sudamérica"
      ) &
      Periodo %in% c("2021S1", "2021S2")
  )
df
```

```r
df$Periodo <- substr(df$Periodo, 1, nchar(df$Periodo) - 2)
```

```r
df2021= df %>% group_by(Nacionalidad) %>% summarise("2021" = sum(Total, na.rm = TRUE))
df2021
```
