```r
df2021= df %>% group_by(Nacionalidad) %>% summarise("2021" = sum(Total, na.rm = TRUE))
df2021
```

```r
# Crear un diccionario de traducción manual
diccionario <- data.frame(
  Nacionalidad_es = c(
    "España", "Francia", "Italia", "Bélgica", "Alemania", "Portugal", "Dinamarca",
    "Argelia", "Argentina", "Bangladesh", "Bolivia", "Brasil", "Bulgaria", "Canadá", 
    "Chile", "China", "Colombia", "Cuba", "Ecuador", "Española", "Estados Unidos de América", "Filipinas", "Finlandia", 
    "Gambia", "Ghana", "Guinea", "Guinea Ecuatorial", "Honduras", "India", 
    "Irlanda", "Lituania", "Mali", "Marruecos", "Mauritania", "Moldavia", 
    "México", "Nicaragua", "Nigeria", "Noruega", "Pakistán", "Paraguay", 
    "Países Bajos", "Perú", "Polonia", "Reino Unido", "República Dominicana", 
    "Rumanía", "Rusia", "Senegal", "Suecia", "Suiza", "Ucrania", 
    "Uruguay", "Venezuela"
  ),
  Nacionalidad_en = c(
    "Spain", "France", "Italy", "Belgium", "Germany", "Portugal", "Denmark",
    "Algeria", "Argentina", "Bangladesh", "Bolivia", "Brazil", "Bulgaria", "Canada", 
    "Chile", "China", "Colombia", "Cuba", "Ecuador", "Spanish", "United States of America", "Philippines", "Finland", 
    "Gambia", "Ghana", "Guinea", "Equatorial Guinea", "Honduras", "India", 
    "Ireland", "Lithuania", "Mali", "Morocco", "Mauritania", "Moldova", 
    "Mexico", "Nicaragua", "Nigeria", "Norway", "Pakistan", "Paraguay", 
    "Netherlands", "Peru", "Poland", "United Kingdom", "Dominican Republic", 
    "Romania", "Russia", "Senegal", "Sweden", "Switzerland", "Ukraine", 
    "Uruguay", "Venezuela"
  )
)

# Realizar el join para reemplazar los nombres en español por los nombres en inglés
df2021 <- df2021 %>%
  left_join(diccionario, by = c("Nacionalidad" = "Nacionalidad_es")) %>%
  mutate(Nacionalidad = ifelse(is.na(Nacionalidad_en), Nacionalidad, Nacionalidad_en)) %>%
  select(-Nacionalidad_en)  # Eliminar la columna auxiliar si no es necesaria

df2021
```

```r
# Merge the coordinates with df2021 based on the country names
df_final <- merge(df2021, coordinates_df, by.x = "Nacionalidad", by.y = "Country", all.x = TRUE)

df_final
```
